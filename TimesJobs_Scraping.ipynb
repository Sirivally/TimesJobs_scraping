```python
from bs4 import BeautifulSoup as bs 
import requests
import pandas as pd 

JS=[]
C=[]
JL=[]
KS=[]
MI=[]

url="https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation="
html=requests.get(url)
content=bs(html.content)
box=content.find("li",class_="clearfix job-bx wht-shd-bx")
boxes=content.find_all("li",class_="clearfix job-bx wht-shd-bx")

for box in boxes:
    Job_skill=box.find("strong",class_="blkclor").text
    company=box.find("h3",class_="joblist-comp-name").text
    Key_Skills=box.find("span",class_="srp-skills").text
    More_info=box.header.h2.a['href']
    
    JS.append(Job_skill)
    C.append(company)
    KS.append(Key_Skills)
    MI.append(More_info)
    
    a=pd.DataFrame(JS,columns=["Job Skill"]) 
    b=pd.DataFrame(C,columns=["Company"])
    c=pd.DataFrame(KS,columns=["key Skills"])
    d=pd.DataFrame(MI,columns=["More Information"])
    
    
    
```

```python
DF=pd.concat([a,b,c,d],axis=1) 
```

```python
DF.to_csv("timejobs.csv")
```
